```python
import pandas as pd
import os
from pathlib import Path

INPUT_DIR = Path.home() / "Desktop" / "newData"

student_info_train = pd.read_csv(os.path.join(INPUT_DIR, "train/student_info.csv"))
student_vle_train = pd.read_csv(os.path.join(INPUT_DIR, "train/student_vle.csv"))
student_reg_train = pd.read_csv(os.path.join(INPUT_DIR, "train/student_registration.csv"))
student_assessment_train = pd.read_csv(os.path.join(INPUT_DIR, "train/student_assessment.csv"))
                                 
student_info_val = pd.read_csv(os.path.join(INPUT_DIR, "val/student_info.csv"))
student_vle_val = pd.read_csv(os.path.join(INPUT_DIR, "val/student_vle.csv"))
student_reg_val = pd.read_csv(os.path.join(INPUT_DIR, "val/student_registration.csv"))
student_assessment_val = pd.read_csv(os.path.join(INPUT_DIR, "val/student_assessment.csv"))
                                 
student_info_test = pd.read_csv(os.path.join(INPUT_DIR, "test/student_info.csv"))
student_vle_test = pd.read_csv(os.path.join(INPUT_DIR, "test/student_vle.csv"))
student_reg_test = pd.read_csv(os.path.join(INPUT_DIR, "test/student_registration.csv"))
student_assessment_test = pd.read_csv(os.path.join(INPUT_DIR, "test/student_assessment.csv"))
                                 
assessments = pd.read_csv("assessments.csv")
courses = pd.read_csv("courses.csv")
vle = pd.read_csv("vle.csv")

# Your dataframes
dfs = {
    "student_info_train": student_info_train,
    "student_vle_train": student_vle_train,
    "student_reg_train": student_reg_train,
    "student_assessment_train": student_assessment_train,
    "student_info_val": student_info_val,
    "student_vle_val": student_vle_val,
    "student_reg_val": student_reg_val,
    "student_assessment_val": student_assessment_val,
    "student_info_test": student_info_test,
    "student_vle_test": student_vle_test,
    "student_reg_test": student_reg_test,
    "student_assessment_test": student_assessment_test,
    "courses": courses,
    "vle": vle,
    "assessments": assessments
}

for name, df in dfs.items():
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    if len(numerical_cols) == 0:
        print(f"{name}: No numerical columns found.")
        continue
    skewness = df[numerical_cols].skew()
    print(f"Skewness for {name}:\n{skewness}\n")
```
