```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

student_info = pd.read_csv("studentInfo.csv")
student_vle = pd.read_csv("studentVle.csv")

total_clicks = student_vle.groupby('id_student')['sum_click'].sum().reset_index()
total_clicks.rename(columns={'sum_click': 'total_clicks'}, inplace=True)
merged_df = pd.merge(student_info, total_clicks, on='id_student', how='left')
merged_df['total_clicks'] = merged_df['total_clicks'].fillna(0)

sns.countplot(data=student_info, x='age_band', hue='final_result')
plt.title('Final Result by Age Band')
plt.xticks(rotation=45)
plt.show()

sns.countplot(data=student_info, x='gender', hue='final_result')
plt.title('Final Result by Gender')
plt.show()

sns.countplot(data=student_info, x='disability', hue='final_result')
plt.title('Final Result by Disability')
plt.show()

sns.countplot(data=student_info, x='imd_band', hue='final_result')
plt.title('Final Result by IMD Band')
plt.xticks(rotation=45)
plt.show()

sns.countplot(data=student_info, x='highest_education', hue='final_result')
plt.title('Final Result by Highest Education')
plt.xticks(rotation=45)
plt.show()

sns.countplot(data=student_info, x='num_of_prev_attempts', hue='final_result')
plt.title('Final Result by Number Of Previous Attempts')
plt.xticks(rotation=45)
plt.show()

sns.countplot(data=student_info, x='studied_credits', hue='final_result')
plt.title('Final Result by Studied Credits')
plt.xticks(rotation=45)
plt.show()

sns.countplot(data=student_info, x='region', hue='final_result')
plt.title('Final Result by Region')
plt.xticks(rotation=45)
plt.show()

print("Final result by total clicks:")
print(merged_df.groupby(['total_clicks', 'final_result']).size())

print("Final result by age band:")
print(student_info.groupby(['age_band', 'final_result']).size())

print("\nFinal result by gender:")
print(student_info.groupby(['gender', 'final_result']).size())

print("\nFinal result by disability:")
print(student_info.groupby(['disability', 'final_result']).size())

print("\nFinal result by imd band:")
print(student_info.groupby(['imd_band', 'final_result']).size())

print("\nFinal result by highest education:")
print(student_info.groupby(['highest_education', 'final_result']).size())

print("\nFinal result by number of previous attempts:")
print(student_info.groupby(['num_of_prev_attempts', 'final_result']).size())

print("\nFinal result by studied credits:")
print(student_info.groupby(['studied_credits', 'final_result']).size())

print("\nFinal result by region:")
print(student_info.groupby(['region', 'final_result']).size())
```
