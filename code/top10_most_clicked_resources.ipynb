```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

student_info = pd.read_csv("studentInfo.csv")
student_vle = pd.read_csv("studentVle.csv")
vle = pd.read_csv("vle.csv")

merged_vle = pd.merge(student_vle, vle, on=['id_site', 'code_module', 'code_presentation'], how='left')

resource_clicks = merged_vle.groupby('activity_type')['sum_click'].sum().sort_values(ascending=False).head(10)

resource_clicks.plot(kind='barh', figsize=(10, 6), legend=False)
plt.title('Top 10 Most Clicked VLE Resource Types')
plt.xlabel('Total Clicks')
plt.ylabel('Activity Type')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

```
